### Decompose the time zones into Windows and non-Windows users

```python
cframe = frame[frame.a.notnull()]
```

```python
print cframe
```

##### Compute a value whether each row is Windows or not

```python
operating_system = np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
```

```python
operating_system[:5]
```

##### Group the data by its time zone column and this new list of operating systems

```python
by_tz_os = cframe.groupby(['tz', operating_system])
```

##### The group counts, analogous to the value_counts() above is computed using size. This is then reshaped into a table with unstack

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
```

```python
agg_counts[:10]
```

##### Select top overall time zones.  Construct an indirect index array from the row counts in agg_counts:

##### Use to sort in ascending order

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

##### Use take to select the rows in that order then slice off the last 10 rows:

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

##### Plot the values in the barplot

```python
count_subset.plot(kind='barh', stacked=True)
```

##### Plot after normalizing the rows to sum to 1

```python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
```

```python
normed_subset.plot(kind='barh', stacked=True)
```
