```python
import numpy as np
import matplotlib.pyplot as pp
```

##### Instruct Ipython notebook to keep pyplots inline in the notebook itself

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,50)
x
```

```python
sinx = np.sin(x)
sinx
```

```python
pp.plot(x,sinx)
```

```python
cosx = np.cos(x)
cosx
```

```python
pp.plot(x,sinx)
pp.plot(x,cosx)
```

```python
pp.plot(x,sinx,'x')
pp.plot(x,cosx,'o')
```

```python
a = sinx*cosx
b = cosx**2 - sinx**2
```

```python
pp.plot(x,a)
pp.plot(x,b)

pp.legend(['sinx(x) cos(x)', 'cos(x)^2 - sin(x)^2]'])
```
